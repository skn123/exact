```python
import struct
import numpy as np
```

```python
def read_bin(filename, verbose):
    max_edge_pheromone = 0
    count=0
    Nodes    = list()
    Edges    = []
    RecEdges = []
    with open(filename, "rb") as f:
        colony_id = struct.unpack("<i",f.read(4))[0]
        print "Colony ID: {}".format(colony_id)
        recurrence_depth = struct.unpack("<i",f.read(4))[0]
        print "Recurrent Depth: {}".format(recurrence_depth)
        no_nodes = struct.unpack("<i",f.read(4))[0]
        print "Number of Nodes in Colony: {}".format(no_nodes)

        for D in range( no_nodes ):
            node_id = struct.unpack("<i",f.read(4))[0]
            print "Node ID: {}".format(node_id)
            no_node_type = struct.unpack("<i",f.read(4))[0]
            print "Number of Node Types: {}".format(no_node_type)
            if node_id!=-1:
                for x in range(no_node_type):
                    node_pheromone = struct.unpack("d",f.read(8))[0]
                    print "Node Pheromone: {}".format(node_pheromone)

            Nodes.append( node_id )

            no_pheromone_lines = struct.unpack("<i",f.read(4))[0]
            for l in range(no_pheromone_lines):
                edge_id = struct.unpack("<i",f.read(4))[0]
                print "Edge ID: {}".format(edge_id)
                edge_pheromone = struct.unpack("d",f.read(8))[0]
                print "Edge Pheromone: {}".format(edge_pheromone)
                edge_pheromone_ = int( (edge_pheromone)/( 2.4 ) * 9 )
                print "Edge Pheromone COLOR: {}".format(edge_pheromone_)
                if max_edge_pheromone<edge_pheromone:
                    max_edge_pheromone = edge_pheromone
                edge_depth = struct.unpack("<i",f.read(4))[0]
                print "Edge Depth: {}".format(edge_depth)
                edge_IN = struct.unpack("<i",f.read(4))[0]
                print "Edge I/P Node: {}".format(edge_IN)
                edge_OUT = struct.unpack("<i",f.read(4))[0]
                print "Edge O/P Node: {}".format(edge_OUT)

                if edge_depth!=0:
                    Edges.append([edge_id, edge_IN, edge_OUT, edge_pheromone])
                else:
                    RecEdges.append([edge_id, edge_IN, edge_OUT, edge_pheromone, edge_depth])

    print "MAX PHEROMONE VALUE: {}".format(max_edge_pheromone)

    
    print Nodes
    Nodes = np.array(Nodes)
    
    Edges = np.array(Edges)
    RecEdges = np.array(RecEdges)
    return Nodes, Edges, RecEdges
```
